# facebox

```python
from PIL import Image
import numpy as np
facebox = []
for i in range(len(img_List)):
    temp_box = detector.detect_faces(np.array(Image.open(img_List[i])))
    facebox.append(temp_box)
    # print(temp_box)
```

```python
aver = np.sum(facebox, axis=0)/len(facebox)
print(aver)
```

```python
%matplotlib inline
import cv2
import matplotlib.pyplot as plt


def showHC(imgpath, facebox):
    font = cv2.FONT_HERSHEY_SIMPLEX

    frame = cv2.imread(imgpath)
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)

    faces = facebox
    print(facebox)

    # Draw a rectangle around the faces
    x, y, w, h = faces
    cv2.rectangle(frame, (x, y), (w, h), (255, 0, 0), 3)
    cv2.putText(frame,'Face',(x, y), font, 2,(255,0,0),5)

    cv2.putText(frame,'Number of Faces : ' + str(len(faces)),(40, 40), font, 1,(255,0,0),3)
    plt.imshow(frame)
    plt.show()

    
aver_face = aver[0][:-1]
aver_face = [int(i) for i in aver_face]
showHC(img_List[0], aver_face)
showHC(img_List[0], [int(i) for i in facebox[0][0][:-1]])
```
